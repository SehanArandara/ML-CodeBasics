```python
import pandas as pd
```

```python
data = pd.read_csv('titanic.csv')
data.head()
```

```python
data.isnull()
```

```python
new_data = data.dropna()
```

```python
new_data.head()
```

```python
new_data.shape
```

```python
new_data.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
```

```python
new_data.head()
```

```python
from sklearn.preprocessing import LabelEncoder

le_sex =  LabelEncoder()
```

```python
new_data['sex_new'] = le_sex.fit_transform(new_data['Sex'])
```

```python
last_data = new_data.drop(['Sex'],axis='columns')
last_data.head()
```

```python
from sklearn import tree

model = tree.DecisionTreeClassifier()
```

```python
from sklearn.model_selection import train_test_split

x = last_data.iloc[:,1:5]
y = last_data.iloc[:,0]
x_test , x_train , y_test , y_train = train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
x_test.shape
```

```python
x_train.shape
```

```python
x.shape
```

```python
model.fit(x_train,y_train)
```

```python
pred = model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(pred,y_test)
cm
```

```python
model.score(x_test,y_test)
```
