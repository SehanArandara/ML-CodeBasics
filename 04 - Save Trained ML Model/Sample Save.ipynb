```python
import numpy as np
import pandas as pd
from sklearn import linear_model
```

```python
df =  pd.read_csv('../01 Linear Regression/homeprices.csv')
df.head()
```

```python
x_area =  df.iloc[:,:1]
y_price = df.iloc[:,1]
```

```python
model = linear_model.LinearRegression() 
model.fit(x_area,y_price)
```

```python
model.predict([[5000]])
```

```python
# we can use pickle to save model 
```

```python
import pickle
```

```python
with open('model_pickle','wb') as f :
    pickle.dump(model,f)
```

```python
# we can load the file like this 
```

```python
with open('model_pickle','rb') as f :
    mp = pickle.load(f)
```

```python
mp.predict([[5000]])
```

```python
# if we have more numpy arrays joblib is more efficient in saving models 
```

```python
import joblib
```

```python
joblib.dump(model,'model_joblib')
```

```python
model_object = joblib.load('model_joblib')
```

```python
model_object.predict([[5000]])
```
