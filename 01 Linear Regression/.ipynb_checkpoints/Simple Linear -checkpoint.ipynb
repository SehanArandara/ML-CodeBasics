```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df =  pd.read_csv('homeprices.csv')
df
```

```python
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price,color='red')
```

```python
x_area =  df.iloc[:,:1]
y_price = df.iloc[:,1]
```

```python
model = linear_model.LinearRegression() 
model.fit(x_area,y_price)
```

```python
model.predict([[3300]])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict([[5000]])
```

## Generates CSV file with list of home price prediction 

```python
area_df = pd.read_csv('areas.csv')
area_df.head()
```

```python
area_pred = model.predict(area_df)
area_pred
```

```python
area_df['price'] = area_pred
area_df
```

```python
area_df.to_csv("prediction.csv")
```
